```python
def inputConv(dBm, amplification): #[Amplification] = dB
    maxP = 1e-3*np.power(10,-6/10) #W
    dBm_val = dBm - amplification
    P = 1e-3*np.power(10,dBm_val/10) #W
    AMP = (P/maxP) #Amplitude of a single peak
    return AMP
```
