```python
import matplotlib.pyplot as plt
import matplotlib
# %matplotlib qt

import numpy as np
from numpy import *
from scipy.fft import fft, ifft, fftshift, ifftshift
from scipy.optimize import curve_fit

import os
import scipy.io as sio

def psd(f, A1, B):
    return (A1/f) + B

def get_fit(x_data, y_data):
    valid = (x_data > 0) & (y_data > 0) & np.isfinite(y_data)
    x_clean = x_data[valid]
    y_clean = y_data[valid]

    params, params_cov = curve_fit(
        psd,
        x_clean,
        y_clean,
#         p0=[1e-6, 1e-6],
#         bounds=(0, np.inf)  # all parameters >= 0
    )

    A1, B = params
    
    return A1, B, x_clean, y_clean
```

```python
initial_folder = r'C:\Users\Demag\PycharmProjects\ColdMeasurements\DeMag\DATA\TWPA_noise_9'

folders = os.listdir(initial_folder)
for i in range(len(folders)):
    if ".fig" in folders[i]: pass
    else: 
        print(r"[%d]" %(i), '----------->', folders[i])
```

```python
file = folders[27]
mat = sio.loadmat(initial_folder+"\\"+file)
print(mat.keys(), '\n')

data_gain = mat['data_gain']
print(data_gain.shape)
volts = mat['volts'][0]
print('DC:', volts.shape)
pump_freqs = mat['pump_freqs'][0]
print('pf:', pump_freqs.shape)
pump_powers = mat['pump_powers'][0]
print('pp', pump_powers.shape)
fft_freq = mat['fft_freq'][0]
print('fft_freq:', fft_freq.shape)


data_mag = mat['data_mag'] # on/off
print(data_mag.shape)
data_pha = mat['data_pha'] # on/off
print(data_pha.shape)
```

```python
data_pha_ref = data_pha[0,0,:,1]
data_pha_mea = data_pha[0,0,:,0]

data_mag_ref = data_mag[0,0,:,1]
data_mag_mea = data_mag[0,0,:,0]

data_mag_mea.shape
```

```python
font = {'size'   : 20}
matplotlib.rc('font', **font)
fig, ax = plt.subplots(2,2, figsize=(16, 10))
# ax[0,0].set_ylabel('S(P)', fontsize=14)
# ax[1].set_ylabel('TWPA', fontsize=14)
# ax[1].set_xlabel('Power, dBm', fontsize=14)

ax[0,0].grid(True, which="both", ls="--", alpha=1)
ax[1,0].grid(True, which="both", ls="--", alpha=1)
ax[0,1].grid(True, which="both", ls="--", alpha=1)
ax[1,1].grid(True, which="both", ls="--", alpha=1)

# ax[0,0].set_ylim(1e-5, 1e-3)
# ax[0,1].set_ylim(1e-6, 1e-4)

a_pha = np.array([])
b_pha = np.array([])
a_mag = np.array([])
b_mag = np.array([])

for i in range(len(pump_powers)):
    if mod(i,1) == 0:
        try:
            a, b, xc, yc = get_fit(fft_freq[2:-1], data_pha_mea[i][2:-1]/100**2*10)
            ax[0,0].loglog(xc, yc, '-', markersize=4, label='{}'.format(np.round(pump_powers[i],2)))

            y_fit = psd(xc, a, b)
            ax[0,0].loglog(xc, y_fit, 'k--', linewidth=0.8)

            a_pha = np.append(a_pha, a)
            b_pha = np.append(b_pha, b)
        except:
            pass

        try: 
            a, b, xc, yc = get_fit(fft_freq[2:-1], data_mag_mea[i][2:-1]/100**2*10)
            ax[0,1].loglog(xc, yc, '-', markersize=4, label='{}'.format(np.round(pump_powers[i],2)))

            y_fit = psd(xc, a, b)
            ax[0,1].loglog(xc, y_fit, 'k--', linewidth=0.8)

            a_mag = np.append(a_mag, a)
            b_mag = np.append(b_mag, b)
        except:
            pass


ax[1,0].plot(pump_powers[:len(a_pha)], a_pha)
ax[1,1].plot(pump_powers[:len(b_pha)], b_pha)

ax[1,0].plot(pump_powers[:len(a_mag)], a_mag)
ax[1,1].plot(pump_powers[:len(b_mag)], b_mag)

# ax[0,0].legend()
```

```python
%matplotlib inline


fig, ax = plt.subplots()

# norm = 'backward' # wrong
norm = 'forward' # correct

N = 101
period_N = 10
test_t = np.linspace(0, period_N, N*period_N+1)
test_signal = np.sin(test_t*2*np.pi)

test_FFT = fft(test_signal, norm = norm)
test_FFT = fftshift(test_FFT)
test_freqs = np.linspace(-N/2, N/2, N*period_N+1)
ax.plot(test_freqs, np.imag(test_FFT))


N = 1001
period_N = 10
test_t = np.linspace(0, period_N, N*period_N+1)
test_signal = np.sin(test_t*2*np.pi)

test_FFT = fft(test_signal, norm = norm)
test_FFT = fftshift(test_FFT)
test_freqs = np.linspace(-N/2, N/2, N*period_N+1)
ax.plot(test_freqs, np.imag(test_FFT))


ax.set_xlim(-5,5)

plt.show();
```
