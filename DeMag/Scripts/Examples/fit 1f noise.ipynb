```python
# pip install mat73

import matplotlib.pyplot as plt
from scipy import *
import os
import scipy.io
from scipy.optimize import curve_fit
import mat73
import numpy as np
import matplotlib

font = {'size'   : 20}
matplotlib.rc('font', **font)
```

```python
adress_file = r"C:\Users\mukha\Documents\Aalto\Data\CoSi res\1fnoise"  #1
files = os.listdir(adress_file)
for i in range(len(files)):
    if ".fig" in files[i]: pass
    else: 
        print(r"[%d]" %(i), '----------->', files[i]) 
```

```python
mat = mat73.loadmat(adress_file+"\\"+files[0])
mat
```

```python
Epsilon = mat['Epsilon']
EpsilonMean = mat['EpsilonMean'] 
Ferr = mat['Ferr'] 
Freq = mat['Freq'] 
spectrum_mean = mat['spectrum_mean'] 



print('Length of Epsilon:', len(Epsilon))
print('Length of EpsilonMean:', len(EpsilonMean))
print('Length of Ferr:', len(Ferr))
print('Length of Freq:', len(Freq))
print('Length of spectrum_mean:', len(spectrum_mean))
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(Freq, spectrum_mean, '.-')
ax.set_yscale("log")
ax.set_xscale("log")
ax.grid(True, which="both", ls="-")
```

```python
def linear_func(x, a, b):
    return a * x + b

def linear_fit(x, y):
    popt, pcov = curve_fit(linear_func, x, y)
    a = popt[0]
    b = popt[1]
    return a, b

start = 2
freq_selected = Freq[start:]
spect_m_selected = spectrum_mean[start:]

a, b = linear_fit(np.log(freq_selected), np.log(spect_m_selected))

fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(Freq, spectrum_mean)
ax.plot(freq_selected, np.exp(linear_func(np.log(freq_selected), a, b)), label=r'$\alpha$ = %1.3f' % np.round(a, 3))
ax.legend()
ax.set_yscale("log")
ax.set_xscale("log")
ax.grid(True, which="both", ls="-")


print('power from fit', a)
```
